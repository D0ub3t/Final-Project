# How can we use chemical composition and crystal structure to predict Moh’s hardness of a mineral?

```python
mineral_data["Mohs Hardness"].corr(mineral_data["Silicon"])
```

```python
mineral_data["Mohs Hardness"].corr(mineral_data["Crystal Structure"])
```

```python
column_names = mineral_data.columns
column_names
```

```python
to_remove = ['Unnamed: 0','Name', 'Crystal Structure', 'Mohs Hardness',
       'Diaphaneity', 'Specific Gravity', 'Optical', 'Refractive Index',
       'Dispersion','count', 'Molar Mass', 'Molar Volume',
       'Calculated Density']
chemical_composition = [item for item in column_names if item not in to_remove]
 #list of all the chemical names
```

```python
chemical_corr_score = mineral_data[chemical_composition].corr(numeric_only=True)
chemical_corr_score #correlation between the chemicals
```

```python
chemical_corr = []
for chemical in chemical_composition:
  correlation = mineral_data["Mohs Hardness"].corr(mineral_data[f"{chemical}"])
  chemical_corr.append(correlation)
 #correlation scores of each chemical and mohs hardness
```

```python
chemical_corr_df = pd.DataFrame({"Chemical": chemical_composition, "Correlation": chemical_corr})
chemical_corr_df #data frame that contains the correlation between each chemical and mohs hardness
```

```python
plt.bar(chemical_corr_df["Chemical"], chemical_corr_df["Correlation"])
plt.xticks(rotation=45)
plt.xlabel("Chemical")
plt.ylabel("Correlation")
plt.title("Correlation of Chemical Composition with Mohs Hardness")
plt.figure(figsize=(50, 30))
```

```python
print(chemical_corr_df.loc[chemical_corr_df['Correlation'].idxmax()])
print(chemical_corr_df.loc[chemical_corr_df['Correlation'].idxmin()])
#these are the most significant chemicals with correlation to mohs hardness
```

```python
#answering the initial question will be the most effective using Hydrated water as an indicator
#the high negative correlation will make it easier to predict the hardness of a mineral
```

```python
model = scipy.stats.linregress(mineral_data["Mohs Hardness"], mineral_data["Hydrated Water"])
plt.scatter(mineral_data["Mohs Hardness"], mineral_data["Hydrated Water"], label = 'Data', color = "#ffd166", marker = 'o')
slope = model.slope
intercept = model.intercept
x = mineral_data["Mohs Hardness"]
fit = slope*x + intercept
plt.plot(mineral_data['Mohs Hardness'], fit, label = 'Line', color ='#ef476f')
plt.xlabel("Mohs Hardness")
plt.ylabel("Hydrated Water")
plt.title("Mohs Hardness vs. Hydrated Water")
plt.legend()
print('Slope:', slope)
#there is a mild correlation between the hydrated water and hardness
#this makes the hydrated water useful in calculating the hardness
```

```python
model = scipy.stats.linregress(mineral_data["Mohs Hardness"], mineral_data["Crystal Structure"])
plt.scatter(mineral_data["Mohs Hardness"], mineral_data["Crystal Structure"], label = 'Data', color = "#db2763", marker = 'o')
slope = model.slope
intercept = model.intercept
x = mineral_data["Mohs Hardness"]
fit = slope*x + intercept
plt.plot(mineral_data['Mohs Hardness'], fit, label = 'Line', color ='#b0db43')
plt.xlabel("Mohs Hardness")
plt.ylabel("Crystal Structure")
plt.title("Mohs Hardness vs. Crystal Structure")
plt.legend()
print('Slope:', slope)
#there is a relatively strong correlation between the crystal structure and hardness
#this makes it a good predictor of the hardness
```

```python
cont_crystal = pd.crosstab(mineral_data["Mohs Hardness"], mineral_data["Crystal Structure"])
cont_water = pd.crosstab(mineral_data["Mohs Hardness"], mineral_data["Hydrated Water"])

chi2_crystal = scipy.stats.chi2_contingency(cont_crystal)
chi2_water = scipy.stats.chi2_contingency(cont_water)

print("Crystal Structure:")
print("Chi-squared statistic:", chi2_crystal[0])
print("p-value:", chi2_crystal[1])

print("Hydrated Water:")
print("Chi-squared statistic:", chi2_water[0])
print("p-value:", chi2_water[1])

#a large chi2 statistic value indicates a significant relationship between the two variables
#crystal structure is much more related than the hydrated water

#the p-values indicate how likely it is that the variable is effecting the data
#it is likly that the hydrated water amount does not relate to the prediction of hardness in a significant way
#we will keep it in the testing for the sake of teh original question
```

```python
#k-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
# Train-test split
from sklearn.model_selection import train_test_split
# Accuracy Score
from sklearn.metrics import accuracy_score
# Confusion Matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
mineral_data['Mohs Hardness'].max() #the data needs to be something other than continous
```

```python
def mapper (x):
  if x==0:
    return 0
  elif x<1:
    return 1
  elif x<2:
    return 2
  elif x<3:
    return 3
  elif x<4:
    return 4
  elif x<5:
    return 5
  elif x<6:
    return 6
  elif x<7:
    return 7
  elif x<8:
    return 8
  elif x<9:
    return 9
  else:
    return 10
```

```python
mineral_data['Mohs Hardness Simple'] = mineral_data['Mohs Hardness'].map(lambda x: mapper(x))
```

```python
#machine learning
X = mineral_data[["Hydrated Water", "Crystal Structure"]]
y = mineral_data["Mohs Hardness Simple"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
#train
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train, y_train)
```

```python
#test
prediction = classifier.predict(X_test)
print(prediction[0:100])
```

```python
#evaluate
accuracy = accuracy_score(y_test, prediction)
print(accuracy*100)
#80% is a good number for the kind of data we have present
```

```python
ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test, cmap="Blues")
```

```python
#the same test performed without Hydrated Water
X = mineral_data[["Crystal Structure"]]
y = mineral_data["Mohs Hardness Simple"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)

classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train, y_train)

prediction = classifier.predict(X_test)
print(prediction[0:100])

accuracy = accuracy_score(y_test, prediction)
print(accuracy*100)
#the accuarcy slightly increased with the removal of the Hydrated Water
#this is consistent with the tests we had previously performed
```

```python
ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test, cmap="summer")
```

```python
plt.scatter(mineral_data['Specific Gravity'], mineral_data['Refractive Index'], marker = '*', color = 'pink')
plt.xlabel('Specific Gravity')
plt.ylabel('Refractive Index')
plt.title('Relationship between Specific Gravity and Refractive Index')
#looking at the graph it can be determined that there is strong positive correlation between the two variables
```

```python
cont_specific_gravity = pd.crosstab(mineral_data["Specific Gravity"], mineral_data["Crystal Structure"])
cont_refractive_index = pd.crosstab(mineral_data["Refractive Index"], mineral_data["Crystal Structure"])

chi2_sg = scipy.stats.chi2_contingency(cont_specific_gravity)
chi2_ri = scipy.stats.chi2_contingency(cont_refractive_index)

print("Specific Gravity:")
print("Chi-squared statistic:", chi2_sg[0])
print("p-value:", chi2_sg[1])

print("Refractive Index:")
print("Chi-squared statistic:", chi2_ri[0])
print("p-value:", chi2_ri[1])

#the below data indicates that there is a very strong relationship between the variables and crystal structure
#the high chi2 values indicate a strong chance the variable is not independent of crystal structure
#the low p-value indicates that the variable is having a strong effect on the structure or vise versa
```

```python
X = mineral_data[["Specific Gravity", "Refractive Index"]]
y = mineral_data["Crystal Structure"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)

classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train, y_train)

prediction = classifier.predict(X_test)

accuracy = accuracy_score(y_test, prediction)
print(accuracy*100)
#an 80% accuracy is reasonable given the type of data and the correlation scores
```

```python
ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test, cmap="autumn")
```

#ANOVA
analysis of variance

```python
import statsmodels.formula.api as sm
from statsmodels.stats.anova import anova_lm
```

```python
mineral_data_adj = mineral_data.rename(columns={'Specific Gravity': 'SpecificGravity'})
mineral_data_adj = mineral_data_adj.rename(columns={'Refractive Index': 'RefractiveIndex'})
mineral_data_adj = mineral_data_adj.rename(columns={'Crystal Structure': 'CrystalStructure'})
```

```python
model = sm.ols('SpecificGravity ~ CrystalStructure', data=mineral_data_adj).fit()
```

```python
anova_results = anova_lm(model)
print(anova_results)
#a large F value indicates that there is a greater difference between the means in the group
#this means that there are real differences between the variables (this is good)
#this confirms our previous statements
```

```python
model = sm.ols('RefractiveIndex ~ CrystalStructure', data=mineral_data_adj).fit()
```

```python
anova_results = anova_lm(model)
print(anova_results)
#a large F value indicates that there is a greater difference between the means in the group
#this confirms our previous statements as well
```
